```python
class EDAPIQuoteWidget():
    _ek = None
    _ric = None
    _widget = None
    _name = None
    _streaming_prices = None
    _dict = None
    def __init__(self, ek, name, dict):
        self._ek = ek
        self._name = name
        self._widget = QuoteWidget()
        self._widget.name = name
        self._dict = dict
        
    def _display_update(self, streaming_prices, instrument_name, fields):        
        self._widget.payload = fields
    
    def _display_refresh(self, streaming_prices, instrument_name, fields):
        fields["X_RIC_NAME"] = instrument_name
        fields["RDN_EXCHD2"] = self._dict.GetExpanedEnumString("RDN_EXCHD2", fields["RDN_EXCHD2"])
        fields["CURRENCY"] = self._dict.GetExpanedEnumString("CURRENCY", fields["CURRENCY"])
        self._display_update(streaming_prices, instrument_name, fields)

    def _display_status(self, streaming_prices, instrument_name, status):        
        self._widget.status = str(status["status"])+" / "+status["message"]
    
    def close(self):
        self._widget.status="Closed"
        self._streaming_prices.close()
        
        
    def open(self, ric):
        self._ric = ric        
        self._streaming_prices = self._ek.StreamingPrices(
            instruments = [self._ric],
            fields = ['X_RIC_NAME',
              'DSPLY_NAME',
              'RDN_EXCHD2',
              'CURRENCY',
              'NETCHNG_1',
              'PCTCHNG',
              'TRDVOL_1',
              'TRADE_DATE',
              'TRDTIM_1',
              'TRDPRC_1',
              'TRDPRC_2',
              'TRDPRC_3',
              'TRDPRC_4',
              'TRDPRC_5', 
              'PRCTCK_1',
              'BID',
              'BIDSIZE',              
              'ASK', 
              'ASKSIZE',
              'OPEN_PRC',
              'ACVOL_1',
              'VWAP',
              'HIGH_1',
              '52WK_HIGH',
              'TURNOVER',
              'LOW_1',
              '52WK_LOW',
              'PERATIO',
              'HST_CLOSE',
              'ADJUST_CLS',
              'EARNINGS',
              'QUOTIM',
              'SEQNUM'],
            on_refresh = lambda streaming_prices, instrument_name, fields:self._display_refresh(streaming_prices, instrument_name, fields),
            on_update = lambda streaming_prices, instrument_name, fields:self._display_update(streaming_prices, instrument_name, fields),
            on_status = lambda streaming_prices, instrument_name, status:self._display_status(streaming_prices, instrument_name, status)
            )
        self._streaming_prices.open()
        
    def widget(self):
        return self._widget
                
```
